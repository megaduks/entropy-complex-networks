```python
%matplotlib inline

import networkx as nx
import numpy as np

np.random.seed(1)

from networkentropy.embeddings import UnoEmbedding, random_walk_generator
from networkentropy.embeddings import DirichletCrossEntropy, CrossEntropy
```

```python
graph = nx.karate_club_graph()

nx.draw(graph, with_labels=True)
```

```python
loss = DirichletCrossEntropy(graph=graph, criterion=nx.betweenness_centrality, alpha=0.025)

model = UnoEmbedding(embedding_size=2,
                     network_size=len(graph.nodes),
                     walk_length=5,
                     num_walks=10,
                     walk_generator=random_walk_generator,
                     loss=loss)

model.fit(graph=graph)
```

```python
layout = nx.get_node_attributes(graph, 'uno')

nx.draw(graph, layout, with_labels=False, edge_color='lightgray', node_size=[v*1000 for k,v in nx.betweenness_centrality(graph).items()])
```

```python

loss = CrossEntropy()

model = UnoEmbedding(embedding_size=2,
                     network_size=len(graph.nodes),
                     walk_length=5,
                     num_walks=10,
                     walk_generator=random_walk_generator,
                     loss=loss)

model.fit(graph=graph)
```

```python
layout = nx.get_node_attributes(graph, 'uno')

nx.draw(graph, layout, with_labels=True)
```
